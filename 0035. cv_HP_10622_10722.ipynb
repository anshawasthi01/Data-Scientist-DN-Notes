```python
# import libraries
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/krishnaik06/Types-Of-Cross-Validation/main/cancer_dataset.csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.drop('Unnamed: 32',axis=1,inplace=True)
```

```python
df.isnull().sum()
```

```python
x = df.iloc[:,2:]
y = df.iloc[:,1]
```

```python
## KFOLD
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import KFold, cross_val_score
k = KFold(5)
model = LogisticRegression()
```

```python
np.mean(cross_val_score(model,x,y, cv = k))
```

```python
model = LogisticRegression()
arr = [2,3,4,5,6]
for i in arr:
    k = KFold(i)
    result = cross_val_score(model,x,y,cv=k)
    print(f'when k={i}',np.mean(result))
    
```

```python
from sklearn.model_selection import StratifiedKFold, cross_val_score
k = StratifiedKFold(n_splits=5)
result = cross_val_score(model,x,y,cv=k)
np.mean(result)
```

```python
## LOOCV
from sklearn.model_selection import LeaveOneOut
lc = LeaveOneOut()
result = cross_val_score(model,x,y,cv=lc)
```

```python
np.mean(result)
```

# HYPER-PARAMETER TUNING

```python
from sklearn import svm, datasets
iris = datasets.load_iris()
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
df
```

```python
df['target'] = iris.target
```

```python
df
```

```python
df['target'] = df['target'].apply(lambda x : iris.target_names[x])
```

```python
df
```

```python
kernels = ['rbf','linear']
C = [1,10,30,20]
avg_scores={}
for kval in kernels:
    for cval in C:
        cv_score = cross_val_score(svm.SVC(kernel= kval, C= cval,gamma='auto'),iris.data,iris.target, cv=5)
        avg_scores[kval + '_' + str(cval)] = np.mean(cv_score)
```

```python
avg_scores
```

```python
from sklearn.model_selection import GridSearchCV
clf = GridSearchCV(svm.SVC(gamma='auto'),{'C':[1,5,7,10],'kernel':['rbf','linear','sigmoid','poly'],
                                         'max_iter':[1,5,8,10,20]},cv=10)
clf.fit(iris.data,iris.target)
clf.cv_results_
```

```python
pd.DataFrame(clf.cv_results_)
```

```python
clf.best_score_
```

```python
clf.best_params_
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python

clf = RandomizedSearchCV(svm.SVC(gamma='auto'),{'C':[1,5,7,10],'kernel':['rbf','linear','sigmoid','poly'],
                                         'max_iter':[1,5,8,10,20]},cv=10,n_iter=5)
clf.fit(iris.data,iris.target)
clf.cv_results_
```

```python
pd.DataFrame(clf.cv_results_)
```

```python
clf.best_params_
```
